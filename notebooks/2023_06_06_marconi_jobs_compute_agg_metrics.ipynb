# Computing aggregation metrics for the jobs

Here i'm using the test dataset

## Reading the test dataset

```python
import pandas as pd

df_jobs = pd.read_csv("../Marconi-test/plugin=job_table/metric=job_info_marconi100/a_0_f12_singlenode_test_exclusive.csv")
df_power = pd.read_csv("../Marconi-test/plugin=ipmi_pub/metric=total_power/power_jobids_test_exclusive.csv")
df_power['node'] = pd.to_numeric(df_power['node'])
df_power['value'] = pd.to_numeric(df_power['value'])
```

## Computing the metrics

Max, mean, median, std, interquartile range (iqr), inter 10-90 percentile range

```python
from scipy.stats import iqr
import numpy as np

#result = [x.values.mean() for x in df_power[df_power["job_id"]==5217003]['value']]
#[5217003, 3861114]

powertrace_jobs = [df_power[df_power["job_id"]==x]['value'].values for x in df_jobs['job_id'].values]
df_jobs["total_power_max_watts"] = [x.max() for x in powertrace_jobs]
df_jobs["total_power_mean_watts"] = [x.mean() for x in powertrace_jobs]
df_jobs["total_power_median_watts"] = [np.median(x) for x in powertrace_jobs]
df_jobs["total_power_std_watts"] = [x.std() for x in powertrace_jobs]
df_jobs["total_power_iqr_watts"] = [iqr(x, rng=(25, 75)) for x in powertrace_jobs]
df_jobs["total_power_1090range_watts"] = [iqr(x, rng=(10, 90)) for x in powertrace_jobs]


"""
df_jobs["total_power_max_watts"] = [df_power[df_power["job_id"]==x]['value'].values.max() for x in df_jobs['job_id'].values]
df_jobs["total_power_mean_watts"] = [df_power[df_power["job_id"]==x]['value'].values.mean() for x in df_jobs['job_id'].values]
df_jobs["total_power_median_watts"] = [np.median(df_power[df_power["job_id"]==x]['value'].values) for x in df_jobs['job_id'].values]
df_jobs["total_power_std_watts"] = [df_power[df_power["job_id"]==x]['value'].values.std() for x in df_jobs['job_id'].values]
df_jobs["total_power_iqr_watts"] = [iqr(df_power[df_power["job_id"]==x]['value'].values, rng=(25, 75))for x in df_jobs['job_id'].values]
df_jobs["total_power_1090range_watts"] = [iqr(df_power[df_power["job_id"]==x]['value'].values, rng=(10, 90))for x in df_jobs['job_id'].values]
"""

df_jobs[["total_power_max_watts",
         "total_power_mean_watts", 
         "total_power_median_watts", 
         "total_power_std_watts", 
         "total_power_iqr_watts", 
         "total_power_1090range_watts"]]
```
